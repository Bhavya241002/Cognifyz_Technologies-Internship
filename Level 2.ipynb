# Task 2 : Cuisine Combination

> ### Identify the most common combinations of cuisines in the dataset.

```python
df['Cuisines'] = df['Cuisines'].astype(str)
```

```python
df['Cuisines'] = df['Cuisines'].str.split(', ')
```

```python
all_cuisines = []

for i in df['Cuisines']:
    all_cuisines.extend(i)
    
cuisine_combinations_count = pd.Series(all_cuisines).value_counts()

print("Most common combinations of cuisines:")
print(cuisine_combinations_count.head())
```

> ### Determine if certain cuisine combinations tend to have higher ratings.

```python
# Explode the 'cuisines' column to create separate rows for each cuisine
exploded_dataset = df.explode('Cuisines')

# Calculate the average rating for each unique cuisine
avg_ratings_by_cuisine = exploded_dataset.groupby('Cuisines')['Aggregate rating'].mean().sort_values(ascending=False)

# Display the results
print("Average Ratings for Each Cuisine:")
avg_ratings_by_cuisine.head()
```

```python
import geopandas as gpd
from shapely.geometry import Point
import matplotlib.pyplot as plt
```

```python
# Create a GeoDataFrame
geometry = [Point(xy) for xy in zip(df['Longitude'], df['Latitude'])]
geo_df = gpd.GeoDataFrame(df, geometry=geometry)
```

```python
# Plot the map
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
fig, ax = plt.subplots(figsize=(10, 6))
world.plot(ax=ax, color='lightgray')
geo_df.plot(ax=ax, markersize=10, color='red', alpha=0.7)
plt.title('Locations of Restaurants on Map')
plt.show()
```

> ### Identify any patterns or clusters of restaurants in specific areas.

```python
from sklearn.cluster import KMeans
```

```python
# Extract latitude and longitude columns
coordinates = df[['Latitude', 'Longitude']]

# Choose the number of clusters (you may need to experiment with this)
num_clusters = 5

# Apply K-means clustering
kmeans = KMeans(n_clusters=num_clusters, random_state=42)
df['cluster'] = kmeans.fit_predict(coordinates)

# Plot the clusters on the map
plt.scatter(df['Longitude'], df['Latitude'], c=df['cluster'], cmap='viridis', alpha=0.5, marker='o')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Clusters of Restaurants')
plt.show()
```

# Task 4 : Restaurant Chains

> ### Identify if there are any restaurant chains present in the dataset.

```python
restaurant_name_counts = df['Restaurant Name'].value_counts()

# Display restaurant names that appear more than once (potential chains)
chains = restaurant_name_counts[restaurant_name_counts > 1]
print("Restaurant Chains in the Dataset:")
print(chains)
```

> ### Analyze the ratings and popularity of different restaurant chains.

```python

# Filter out rows with null values in 'Rating' and 'Votes'
df = df.dropna(subset=['Aggregate rating', 'Votes'])

# Group by 'Restaurant Name' to calculate average rating and total votes
chain_stats = df.groupby('Restaurant Name').agg({'Aggregate rating': 'mean', 'Votes': 'sum'})

# Sort by average rating or total votes, depending on your analysis goal
sorted_chain_stats_by_rating = chain_stats.sort_values(by='Aggregate rating', ascending=False)
sorted_chain_stats_by_votes = chain_stats.sort_values(by='Votes', ascending=False)

# Display the results
print("Top Chains by Average Rating:")
print(sorted_chain_stats_by_rating.head())

print("\nTop Chains by Total Votes:")
print(sorted_chain_stats_by_votes.head())
```
